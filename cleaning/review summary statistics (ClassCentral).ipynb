```python
import csv
import datetime
```

```python
## read course Data set
courseList = []
with open("./MOOC_data/xtData/(cleand)ClassCentral_course.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        courseList.append(row)
        
print(len(courseList))
```

```python
## read review Data set
reviewList = []
with open("./MOOC_data/xtData/(Date_Adjusted)ClassCentral_review_renewed.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        reviewList.append(row)
print(len(reviewList))
```

```python
## review statistics init
statistics = []
for t in range(1,5):
    statistics.append(dict({
        "time": t,
        "num_review": 0,
        "num_review_with_id":0,
        "num_low_star":0,
        "num_mid_star":0,
        "num_high_star":0,
        "num_helpful_review":0,
        "sum_review_length":0,
        "sum_review_rating":0,
        "courses": [],
        "num_courses": 0,    
        
    }))
```

```python
for review in reviewList:
    
    temp_review_value = int(review['single_rating_value'])
    date = datetime.datetime.strptime(review['review_date'], "%Y-%m-%d") 
    
    if date.year == 2013:
        # T1: 
        if date.month > 6 and date.month <= 12:
            statistics[0]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[0]['num_review_with_id'] += 1
                
            statistics[0]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[0]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[0]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[0]['num_mid_star'] += 1
            else:
                statistics[0]['num_high_star'] += 1
                
            if review['url'] not in statistics[0]['courses']:
                statistics[0]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[0]['num_helpful_review'] += 1
                

    elif date.year == 2014:

        if date.month > 0 and date.month < 7:
            statistics[0]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[0]['num_review_with_id'] += 1
                
            statistics[0]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[0]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[0]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[0]['num_mid_star'] += 1
            else:
                statistics[0]['num_high_star'] += 1
                
            if review['url'] not in statistics[0]['courses']:
                statistics[0]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[0]['num_helpful_review'] += 1
        # T2: 
        else:
            statistics[1]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[1]['num_review_with_id'] += 1
                
            statistics[1]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[1]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[1]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[1]['num_mid_star'] += 1
            else:
                statistics[1]['num_high_star'] += 1
            
            if review['url'] not in statistics[1]['courses']:
                statistics[1]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[1]['num_helpful_review'] += 1
                
    elif date.year == 2015:
        
        if date.month > 0 and date.month < 7:
            statistics[1]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[1]['num_review_with_id'] += 1
                
            statistics[1]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[1]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[1]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[1]['num_mid_star'] += 1
            else:
                statistics[1]['num_high_star'] += 1
                
                
            if review['url'] not in statistics[1]['courses']:
                statistics[1]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[1]['num_helpful_review'] += 1
                
        # T3: 
        else:
            statistics[2]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[2]['num_review_with_id'] += 1
                
            statistics[2]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[2]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[2]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[2]['num_mid_star'] += 1
            else:
                statistics[2]['num_high_star'] += 1
                
            if review['url'] not in statistics[2]['courses']:
                statistics[2]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[2]['num_helpful_review'] += 1
                
    elif date.year == 2016:

        if date.month > 0 and date.month < 7:
            statistics[2]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[2]['num_review_with_id'] += 1
                
            statistics[2]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[2]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[2]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[2]['num_mid_star'] += 1
            else:
                statistics[2]['num_high_star'] += 1
                
            if review['url'] not in statistics[2]['courses']:
                statistics[2]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[2]['num_helpful_review'] += 1
                
        # T4: 
        else:
            statistics[3]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[3]['num_review_with_id'] += 1
                
            statistics[3]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[3]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[3]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[3]['num_mid_star'] += 1
            else:
                statistics[3]['num_high_star'] += 1
                
            if review['url'] not in statistics[3]['courses']:
                statistics[3]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[3]['num_helpful_review'] += 1
                
    elif date.year == 2017:

        if date.month > 0 and date.month < 7:
            statistics[3]['num_review'] += 1
            if(review['reviewer_url'] != "null"):
                statistics[3]['num_review_with_id'] += 1
                
            statistics[3]['sum_review_rating'] += int(review['single_rating_value'])
            statistics[3]['sum_review_length'] += int(len(review['review_contents']))
                        
            if temp_review_value > 0 and temp_review_value < 3:
                statistics[3]['num_low_star'] += 1
            elif temp_review_value == 3:
                statistics[3]['num_mid_star'] += 1
            else:
                statistics[3]['num_high_star'] += 1
                
            if review['url'] not in statistics[3]['courses']:
                statistics[3]['courses'].append(review['url'])
                
            if int(review['helpful_rate']) > 0:
                statistics[3]['num_helpful_review'] += 1
```

```python
for stat in statistics:
    stat['num_courses'] = len(stat['courses'])
        
```

```python
with open('./MOOC_data/xtData/xt_1year(class_central)/review_data_summary.csv', 'w', newline = '') as write_file:
    
    fieldnames = ['time', 'num_review', 'num_review_with_id', 'frac_low_star', 'frac_mid_star', 'frac_high_star', 'avg_review_len',
                 'avg_review_rating', 'num_reviewed_courses', 'frac_helpful_review']
    
    dict_writer = csv.DictWriter(write_file, fieldnames)
    dict_writer.writeheader()
    
    for stat in statistics:
        temp_dict = {
            'time': stat['time'],
            'num_review': stat['num_review'],
            'num_review_with_id': stat['num_review_with_id'],
            'frac_low_star': stat['num_low_star']/stat['num_review'],
            'frac_mid_star': stat['num_mid_star']/stat['num_review'],
            'frac_high_star': stat['num_high_star']/stat['num_review'],
            'frac_helpful_review': stat['num_helpful_review']/stat['num_review'],
            'avg_review_len': stat['sum_review_length']/stat['num_review'],
            'avg_review_rating': stat['sum_review_rating']/stat['num_review'],
            'num_reviewed_courses': stat['num_courses']                
        }
        dict_writer.writerow(temp_dict)
```
